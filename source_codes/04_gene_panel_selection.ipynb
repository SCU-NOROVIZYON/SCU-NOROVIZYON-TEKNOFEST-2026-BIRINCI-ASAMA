```python
import pandas as pd
import numpy as np

df = pd.read_csv("variant_dataset.csv", low_memory=False)
```

```python
df['genename'] = df['genename'].apply(
    lambda x: ';'.join(sorted(set(x.split(';'))))
)

df = df[df['genename'].apply(lambda x: ';' not in x)]
```
